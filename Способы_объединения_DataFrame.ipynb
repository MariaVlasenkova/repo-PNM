```python
import numpy as np
import pandas as pd
```

```python
zoo = pd.DataFrame([['elephant', 500], ['tiger', 300], ['zebra',200], ['lion', 420], ['kangaroo', 410]], columns=['animal', 'water_need'])
zoo
```

```python
zoo_eats = pd.DataFrame([['elephant','vegetables'], ['tiger','meat'], ['kangaroo','vegetables'], ['zebra','vegetables'], ['giraffe','vegetables']], columns=['animal', 'food'])
zoo_eats
```

#merge

```python
odin = zoo.merge(zoo_eats)
odin
```

#INNER
При выборе INNER JOIN объединяются только те значения, которые можно найти в обеих таблицах.

```python
dva = zoo.merge(zoo_eats, how='inner')
dva
```

# OUTER
В случае же с OUTER JOIN объединяются все значения, даже если некоторые из них есть только в одной таблице.

```python
tri = zoo.merge(zoo_eats, how='outer')
tri
```

#LEFT
Left заберет все значения из левой таблицы (zoo), но из правой (zoo_eats) использует только те значения, которые есть в левой.

```python
chetre = zoo.merge(zoo_eats, how='left')
chetre
```

#RIGHT
Right заберет все значения из правой таблицы (zoo_eats), но из левой (zoo) использует только те значения, которые есть в правой.

```python
pat = zoo.merge(zoo_eats, how='right')
pat
```
